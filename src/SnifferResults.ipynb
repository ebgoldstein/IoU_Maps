```python
#load L1
df1 = pd.read_csv('../data/SnifferResults/L1_1.csv')
df2 = pd.read_csv('../data/SnifferResults/L1_2.csv')

#correct mistake in Sniffer scale
df1['Sorted'] = df1['Sorted'] + 1

L1 = pd.concat([df1, df2], axis=0)
L1.insert(0, 'labeler', '1')
print(L1.shape)
L1.head()
```

```python
#load L2
df1 = pd.read_csv('../data/SnifferResults/L2_1.csv')
df2 = pd.read_csv('../data/SnifferResults/L2_2.csv')


L2 = pd.concat([df1, df2], axis=0)
L2.insert(0, 'labeler', '2')
print(L2.shape)
L2.head()
```

```python
#load L3
df1 = pd.read_csv('../data/SnifferResults/L3_1.csv')
df2 = pd.read_csv('../data/SnifferResults/L3_2.csv')


L3 = pd.concat([df1, df2], axis=0)
L3.insert(0, 'labeler', '3')
print(L3.shape)
L3.head()
```

```python
#load L4
df1 = pd.read_csv('../data/SnifferResults/L4_1.csv')
df2 = pd.read_csv('../data/SnifferResults/L4_2.csv')


L4 = pd.concat([df1, df2], axis=0)
L4.insert(0, 'labeler', '4')
print(L4.shape)
L4.head()
```

```python
#load L5
df1 = pd.read_csv('../data/SnifferResults/L5_1.csv')
df2 = pd.read_csv('../data/SnifferResults/L5_2.csv')


L5 = pd.concat([df1, df2], axis=0)
L5.insert(0, 'labeler', '5')
print(L5.shape)
L5.head()
```

```python
All = pd.concat([L1, L2,L3,L4,L5], axis=0)
print(All.shape)
All.head()
```

```python
pALL = All.pivot(index='Filename', columns='labeler', values='Sorted')
pALL.head()
```

```python
# save this dataframe
pALL.to_csv('../data/SnifferResults/ALL.csv')
```
